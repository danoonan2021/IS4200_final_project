Steps:

1. lowercase everything
2. remove punctuation
3. not sure if lemmatization/stemming is useful since we'd often want exact matches on lyrics

```python
from string import punctuation

def process(lyric):
  processed_lyric = lyric.lower()
  processed_lyric = processed_lyric.translate(str.maketrans('', '', punctuation))
  return processed_lyric
```

```python
data['processed'] = data['lyrics'].apply(process)
```

```python
def simple_search_cleaned(query):
  processed_query = process(query)
  return data[data['processed'].str.contains(processed_query)]
```

```python
from google import genai
```

```python
client = genai.Client(api_key=google_api_key)

def generate_embeddings(text):
  result = client.models.embed_content(
        model="gemini-embedding-exp-03-07",
        contents="What is the meaning of life?")
  return result.embeddings[0]
```

```python
generate_embeddings(data['processed'].iloc[100])
```

```python
data['embedded'] = data['processed'].apply(generate_embeddings)
```

```python
from FlagEmbedding import BGEM3FlagModel

model = BGEM3FlagModel('BAAI/bge-m3',
                       use_fp16=True) # Setting use_fp16 to True speeds up computation with a slight performance degradation
```

```python
model.target_devices
```

```python
docs = data['processed'].tolist()
embeddings = model.encode(docs,
                          batch_size=16,
                          max_length=1024)['dense_vecs']
data['embedded'] = embeddings
```

```python
data['embedded'] = list(embeddings)
```

```python
# save the data to a csv and download
data.to_csv('data.csv', index=False)
```

```python
# define cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def find_most_similar(lyrics):
  lyric_embedded = model.encode(lyrics)['dense_vecs'].reshape(1,-1)
  embedded_lyrics = data['embedded'].tolist()
  similarities = cosine_similarity(lyric_embedded, embedded_lyrics)
  # return top 10
  top_10_indices = similarities.argsort()[0][-10:][::-1]
  return data.iloc[top_10_indices]

```
